```python
# 점수 주는 함수

def calculate_investment_score(investment_list, scores):
    score = 0
    # 위에서 설명한 방식 쓸데없는 str 바꿔서 리스트에 추가
    investment_l = investment_list.replace("[", "").replace("]", "").replace("'", "").replace(" ", "").split(",")
    for investment in investment_l:
        # 리스트 하나하나 for문 돌면서 점수 계산, socres 딕셔너리에 없으면 1로 설정
        score += scores.get(investment, 1)
    return score
```

```python
# 투자점수 컬럼 생성 함수는 calculate_investment_score로 기준은 scores 딕셔너리 변수로
df['투자점수'] = df['투자사'].apply(lambda x: calculate_investment_score(x, scores))
```

```python
# 기업점수 있는 파일과, 연매출 있는 파일 불러오기

df1 = pd.read_csv("final_all_data_end.csv", encoding = 'cp949')
df2 = pd.read_csv("final_investment_company_all_0815.csv", encoding = 'utf-8')
```

```python
# df1 과 df2 병합 기준은 기업명으로 왼쪽병합, df2가 df1으로 병합

merged_df = pd.merge(df1, df2[['기업명', '투자점수']], on='기업명', how='left')
```

```python
# 병합한 파일 기업명 중복된 데이터 확인
merged_df.duplicated('기업명').sum()
```

```python
# 데이터 결측치 제거
clean_df = merged_df.dropna()
```

```python
# 기업명이 중복된 데이터 제거
drop_clean_df = clean_df.drop_duplicates(subset = '기업명')
```

```python
# 최종 파일 저장
drop_clean_df.to_csv("end_all_data_0815.csv", encoding = 'utf-8', index = False)
```
