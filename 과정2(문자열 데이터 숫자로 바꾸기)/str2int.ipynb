```python
# 특허기술 평균등급 "-" 값 0으로 바꾸기

df["특허기술 평균등급"] = df["특허기술 평균등급"].replace('-', 0)
df['특허기술 평균등급'] = df['특허기술 평균등급'].astype(float)
```

```python
# 문자열을 숫자로 변환하는 함수 (만원 억원만)
# 돈
def parse_money_to_number(money_str):
    # 숫자와 . 을 제외한 것은 필터시켜줌
    cle_str = ''.join(filter(lambda x: x.isdigit() or x == '.' or x == '-', money_str))
    #print(money_str, cle_str)
    if money_str == '-':
        return 0
    if '만원' in money_str:
        return int(float(cle_str) * 1)
    elif '억원' in money_str:
        return int(float(cle_str) * 10000)
    elif '조원' in money_str:
        return int(float(cle_str) * 100000000)
    elif '경원' in money_str:
        return int(float(cle_str) * 1000000000000)
    else:
        return int(float(cle_str))

#어떤 컬럼을 변환할지 알려주는 함수 (돈을 수치형으로)
def money_to_int(col):
    # apply() 메서드에 인자 없이 함수를 전달하면
    # 해당 함수를 각 열의 모든 값에 적용함
    df[col+'(만원)'] = df[col].apply(parse_money_to_number)
    df.drop(col, axis = 1, inplace = True)

# 문자열을 숫자로 변환하는 함수 (만, 억)
# 최근투자유치
def parse_invest_to_number(invest_str):
    # 숫자와 . 을 제외한 것은 필터시켜줌
    cle_str = ''.join(filter(lambda x: x.isdigit() or x == '.' or x == '-', invest_str))
    #print(invest_str, cle_str)
    if invest_str == '-':
        return 0
    if invest_str == "비공개":
        return 0
    if '만' in invest_str:
        return int(float(cle_str) * 1)
    elif '억' in invest_str:
        return int(float(cle_str) * 10000)
    else:
        return int(float(cle_str))

def invest_to_int(col):
    # apply() 메서드에 인자 없이 함수를 전달하면
    # 해당 함수를 각 열의 모든 값에 적용함
    df[col+'(만원)'] = df[col].apply(parse_invest_to_number)
    df.drop(col, axis = 1, inplace = True)

# 인원 수
def parse_employees_to_number(employees_str):
    cle_str = ''.join(filter(lambda x: x.isdigit() or x == '.', employees_str))
    return int(float(cle_str))

# 사람을 수치형으로
def people_to_int(col):
    # apply() 메서드에 인자 없이 함수를 전달하면
    # 해당 함수를 각 열의 모든 값에 적용함
    df[col] = df[col].apply(parse_employees_to_number)
    #df.drop(col, axis = 1, inplace = True)
```

```python
money_l = ['자본금', '누적투자유치금액', '연매출',
           '매출액_2020', '매출액_2021', '영업이익_2020', '영업이익_2021',
           '순이익_2020', '순이익_2021', '자산_2020', '자산_2021',
           '자본_2020', '자본_2021', '2022기준 시장규모']
money_2_l = ['최근투자유치']
people_l = ['고용인원', '연간 퇴사율']

all_col = money_l + money_2_l + people_l

for col in all_col:
    if col in money_l:
        money_to_int(col)
    elif col in money_2_l:
        invest_to_int(col)
    else:
        people_to_int(col)
    #print(col)
```

날짜 변환

```python
from datetime import datetime

# 문자열인 설립일자를 날짜 객체형태로 변환
def parse_founding_date(date_str):
    return datetime.strptime(date_str, '%Y-%m-%d')

# 두날짜의 차이를 구하는 함수
def calculate_years_diff(founding_date):
    today = datetime.today()
    diff = today.year - founding_date.year
    return diff
```

```python
# 설립일자 컬럼을 날짜 형식으로 변환
df['설립일자'] = df['설립일자'].apply(parse_founding_date)

# 년차로 변환
df['년차'] = df['설립일자'].apply(calculate_years_diff)

df.drop('설립일자', axis = 1, inplace = True)
```

주소 변환

```python
# 주소데이터에 대전광역시라고 써있는게 있고 대전이라고 써있는게 있었음
# 그래서 서울 대전 제주 이런식으로 규칙을 맞춰줌
def omit_address(adress_str):
    cle_l = adress_str.split(" ")
    if cle_l[0] == "서울시":
        cle_l[0] = "서울"
    elif cle_l[0] == "대전광역시":
        cle_l[0] = "대전"
    elif cle_l[0] == "서울특별시":
        cle_l[0] = "서울"
    elif cle_l[0] == "제주특별자치도":
        cle_l[0] = "제주"
    elif cle_l[0] == "부산광역시":
        cle_l[0] = "부산"
    elif cle_l[0] == "강원특별자치도":
        cle_l[0] = "강원"
    elif cle_l[0] == "인천광역시":
        cle_l[0] = "인천"

    cle_str = ' '.join([cle_l[0], cle_l[1]])
    return cle_str

# 주소 변환
def address_to_add(col):
    # apply() 메서드에 인자 없이 함수를 전달하면
    # 해당 함수를 각 열의 모든 값에 적용함
    df[col] = df[col].apply(omit_address)
```
